# K-means кластеризация
В качестве примера будет использован датасет клиенотов торгового центра

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv('Mall_Customers.xls')
df = df.set_index('CustomerID')
df
```

```python
df.info()
```

```python
df = df.replace({'Male': 1, 'Female': 0})
df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.iloc[:,2],df.iloc[:,3])
```

```python
clustering_score = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random', random_state = 42)
    kmeans.fit(df)
    clustering_score.append(kmeans.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range(1, 11), clustering_score)
plt.scatter(5,clustering_score[4], s = 200, c = 'red', marker='*')
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('Clustering Score')
plt.show()
```

```python
df_test = df[['Annual Income (k$)', 'Spending Score (1-100)']]

clustering_score = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random', random_state = 42)
    kmeans.fit(df_test)
    clustering_score.append(kmeans.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range(1, 11), clustering_score)
plt.scatter(5,clustering_score[4], s = 200, c = 'red', marker='*')
plt.xlabel('No. of Clusters')
plt.ylabel('Clustering Score')
plt.show()
```

```python
c_num = 9
kmeans = KMeans(n_clusters = c_num, init='random', random_state = 42)
kmeanspp = KMeans(n_clusters = c_num, random_state = 42)

kmeans.fit(df_test)
kmeanspp.fit(df_test)

pred = kmeans.predict(df_test)
predpp = kmeanspp.predict(df_test)
```

```python
res = df.copy()
respp = df.copy()

res['Clusters'] = pred
respp['Clusters'] = predpp
```

```python
plt.figure(figsize=(10,6))

tmp = res[['Annual Income (k$)','Spending Score (1-100)']]

for i in range(c_num):
    cl = tmp[res.Clusters == i]
    plt.scatter(cl.iloc[:, 0], cl.iloc[:, 1], label = f'Cluster {i}')
    

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:, 1],s = 300, label = 'Centroid km', marker='*')
plt.scatter(kmeanspp.cluster_centers_[:,0], kmeanspp.cluster_centers_[:, 1],s = 300, label = 'Centroid kmpp', marker='*')

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.title('Customer Clusters')
```

```python
plt.figure(figsize=(10,6))

tmp = respp[['Annual Income (k$)','Spending Score (1-100)']]

for i in range(c_num):
    cl = tmp[respp.Clusters == i]
    plt.scatter(cl.iloc[:, 0], cl.iloc[:, 1], label = f'Cluster {i}')
    

plt.scatter(kmeanspp.cluster_centers_[:,0], kmeanspp.cluster_centers_[:, 1],s = 300, label = 'Centroid kmpp', marker='*')

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.title('Customer Clusters')
```
