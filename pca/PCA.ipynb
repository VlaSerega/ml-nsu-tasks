# PCA
В качестве примера возьмем датасет чего-то

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv").drop(['ID', 'TARGET'], axis=1)
df
```

```python
sns.heatmap(df.corr(), annot=True,cmap='RdYlGn') 
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df)



explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(6, 4))

    plt.bar(range(len(df.columns)), explained_variance, alpha=0.5, align='center',
            label='individual explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```

```python
from sklearn.preprocessing import StandardScaler

df_tr = StandardScaler().fit_transform(df)
```

```python
pca = PCA(n_components = 2)
df_new = pca.fit_transform(df_tr)
```

```python
plt.figure(figsize=(20,12))
sns.scatterplot(x=df_new[:,0], y=df_new[:,1], color="crimson")
plt.show()
```

```python
from umap import UMAP

umap = UMAP(random_state=42)
embedding = umap.fit_transform(df)

plt.figure(figsize=(20,12))
sns.scatterplot(x=embedding[:,0], y=embedding[:,1], color="crimson")
plt.show()
```
